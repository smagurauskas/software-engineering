### What is UI and what is UX

User Interface (UI) focuses on how a product/web looks and interacts visually.

User Experience (UX) focuses on how a product feels and functions.

### UX vs UI

Purpose: UX solves user needs; UI creates visual appeal.

Focus: UX is functionality centered; UI is appearance centered.

Order: UX often comes first (foundation), and UI builds on top (decoration).

## UX

UX design is about solving problems for users.

UX involves researching user needs, organizing content logically, and designing interactions that create a friendly user experience.
